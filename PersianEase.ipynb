```python
from datasets import load_dataset

dataset = load_dataset('csv', data_files='/content/drive/MyDrive/exportStatements.csv')
```

```python
from datasets import DatasetDict

dataset = dataset['train'].train_test_split(test_size=0.2, seed=42)
dataset = DatasetDict({
    'train': dataset['train'].train_test_split(test_size=0.125, seed=42)['train'],
    'validation': dataset['train'].train_test_split(test_size=0.125, seed=42)['test'],
    'test': dataset['test']
})
```

```python
from transformers import (T5ForConditionalGeneration, AutoTokenizer, pipeline)
```

```python
model_path = 'erfan226/persian-t5-formality-transfer'
model = T5ForConditionalGeneration.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
pipe = pipeline(task='text2text-generation', model=model, tokenizer=tokenizer)
```

```python
def preprocess_function(examples):
    inputs = [f'formal: {f}' for f in examples['FormalForm']]
    targets = [f'{f}' for f in examples['inFormalForm']]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding='max_length')

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True, padding='max_length')

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs



tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import TrainingArguments, Trainer, EarlyStoppingCallback

training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/model_checkpoints",  # Change this to your Google Drive directory
    evaluation_strategy = "epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    weight_decay=0.01,
    save_strategy = "epoch",
    load_best_model_at_end=True,
    metric_for_best_model="loss",
    save_total_limit = 1,  # Only the last 1 model are saved. Older ones are deleted.
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)],
)

trainer.train()
```
